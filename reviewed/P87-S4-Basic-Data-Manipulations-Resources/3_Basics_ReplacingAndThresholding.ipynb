# Basics - Replacing and Thresholding

Also super common for cleaning datasets.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("AB_NYC_2019.csv")
df.head(2)
```

## Dealing with NaNs

```python
df.info()
```

```python
#df.dropna(thresh=3)
#axis= 0 == line , axis=1 == row
df.dropna(subset=["last_review"]).info()
```

```python
#You can use one of the methods like backfill or forward fill,

#return to a data frame where every single not a number of value is instead zero.
df.fillna(0);
```

## Generic replace

```python
df.replace("John", "Jono").head(1)
```

```python
#limit to 1 replacement

df.host_name.replace("John", "Jono", limit=1)[0]
```

```python
df.replace({"John": "Jono", "Brooklyn": "Brooky"}).head(1)
```

## Thresholding

```python
import matplotlib.pyplot as plt
#plt.hist(df.price);
plt.hist(df.price, log= True);
```

```python
plt.hist(df.price.clip(upper=1000));
```

```python
df2 = df.copy()
df2.loc[df2.price > 1000, "price"] = 1000
plt.hist(df2.price);
```
