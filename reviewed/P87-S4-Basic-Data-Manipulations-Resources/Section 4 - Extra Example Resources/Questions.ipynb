```python
# Code to start you off and manipulate the data. .T is transpose - swap columns and rows
import pandas as pd

df = pd.read_csv("cost-of-living.csv", index_col=0).T.reset_index()
df2 = df.copy()
#df.head(2)
```

## Rename column  v

```python
#POSSIBILITYS

#df2 = df2.rename({'index' :'location'}, axis=1)  # new method
#df2 = df2.rename({'index' :'location'}, axis='columns')
#df2 = df2.rename(columns={'index' :'location'})

df2 = df.rename(columns={'index' :'location'})
```

## Get city and country

```python
# n (columns ) = lenght keys , as [ 'City', 'Country'] ==  n 1
df2[['City', 'Country']] = df2.location.str.split(",", n= 1 ,expand=True)
```

```python
df2.Country = df2.Country.apply(lambda x: x if not ',' in x else x.split(',')[-1]) # check -1
```

## Figure out which country has the most cities

```python
df2.Country.value_counts()
```

## Create a subset of only that country

```python
most_cities = df2.Country.value_counts().index[0]

df3 = df2[df2.Country == most_cities]
df3.columns.all
```

## Sort by housing accommodation

```python
col = "Apartment (1 bedroom) in City Centre"
#df2.sort_index(['Apartment (1 bedroom) in City Centre'])
df4 = df3.sort_values(col, ascending = False)

df4[['City', col]]
```
